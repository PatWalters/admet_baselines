This notebook provides a few quick examples of how to train the `PXRRegressor` class.

```python
from pxr_regressor import PXRRegressor
import pandas as pd
```

Read the PXR data from the ChEMBL database.

```python
df = pd.read_csv('PXR_chembl_EC50.csv')
```

Train a PXR regressor, note that this repo has an already trained model in the file `PXR_regressor.pkl`.

```python
reg = PXRRegressor()
reg.fit(df['OPENADMET_CANONICAL_SMILES'], df['PXR_pEC50'])
```

Make predictions for the training data.  This is not realistic; it's just an example. 

```python
y_pred, y_std = reg.predict(df['OPENADMET_CANONICAL_SMILES'])

df['PXR_pEC50_pred'] = y_pred
df['PXR_pEC50_std'] = y_std
```

Save the regressor to disk. 

```python
reg.save('PXR_regressor.pkl')
```

Load a regressor from a saved model.

```python
reg2 = PXRRegressor('PXR_regressor.pkl')
```

Make a prediction based on the saved model.

```python
y_pred, y_std = reg2.predict(df['OPENADMET_CANONICAL_SMILES'])

df['PXR_pEC50_pred2'] = y_pred
df['PXR_pEC50_std2'] = y_std
```
